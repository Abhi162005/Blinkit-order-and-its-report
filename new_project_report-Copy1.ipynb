#### Data Cleaning

#### Used functions
- **dropna()**           **-** Used to remove an entire row from dataset if there is any null value in that row(even at in one place).
- **fillna()**           **-** Used to fill the null values with value provided inside of parathesis in the dataset.
- **drop_duplicates()**  **-** Used to find and remove all duplicate rows in the dataset.
- **replace()**          **-** Used to replace the old value with the new one.
- **loc[]**              **-** Used to find and modify values in the rows and columns of the dataset.
- **drop()**             **-** Used to remove the rows or columns from the dataset.

```python
before = len(df)            # removes entire row if any null value found
df.dropna(inplace=True)
after = len(df)
if before > after:
    print(f"{before - after} row(s) with null values dropped.")
else:
    print(" No null values found. No changes made.")
```

```python
# fills all the null values if found with median of it's column
nulls_before = df['customer_id'].isnull().sum()
df.fillna(df['customer_id'].median(), inplace=True)
nulls_after = df['customer_id'].isnull().sum()
if nulls_before > 0:
    print(f"{nulls_before - nulls_after} null values filled with median of the 'customer_id' column.")
else:
    print("No null values found in 'customer_id'. No changes made.")
```

```python
# drops the duplicate rows from the dataset 
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)
if before > after:
    print(f"{before - after} duplicate rows removed.")
else:
    print("No duplicate rows found. No changes made.")
```

## Visualizations

### 1.) MOST SOLDOUT ITEM ANALYSIS

### 2.) MOST USED PAYMENT OPTION

```python
# PAYMENT METHOD ANALYSIS
counts = df['payment_method'].value_counts()
counts.plot(kind='bar', color='skyblue')
plt.title('Payment Method Count')
plt.xlabel('Payment Method')
plt.ylabel('Count')
for i, value in enumerate(counts):
    plt.text(i, value + 1, str(value), ha='center')
plt.tight_layout()
plt.show()
```

### 3.) NO.OF PRODUCTS IN EACH CATEGORY

```python
# PRODUCT'S SALES DISTRIBUTION AND ANALYSIS
df.groupby('category')['product_name'].nunique().sort_values().plot(kind='barh', title='Unique Products per Category')
plt.ylabel('Unique Products Category')
plt.xlabel('No.of types Items in that category')
plt.tight_layout()
plt.show()
```

### 4.) PRICE DISTRIBUTION 
1. Median (Q2) – the line inside the box
    This is the middle value of the prices.
2. First Quartile (Q1) – the bottom of the box
    25% of the data is below this value.
3. Third Quartile (Q3) – the top of the box
    75% of the data is below this value.
4. Interquartile Range (IQR) – the height of the box
    It’s the range between Q3 and Q1. It shows the middle 50% of the data.
5. Whiskers – the lines extending from the box
    These typically show the range of values within 1.5 × IQR from the quartiles. They help you understand the spread.
6. Outliers – the dots beyond the whiskers
### Example Interpretation:
If the box plot for price looks like this:
 ● The box is tall → prices vary a lot.
 ● Many outliers on top → some products are much more expensive than most.
 ● Short box → most prices are close to each other.
 ● Skewed box (if median is not centered) → data is skewed (e.g., more low-priced or high-priced items).

```python
df['unit_price'].plot(kind='box', title='Price Distribution')
plt.ylabel('Price')
plt.show()
```

### 5.) Monthly revenue generated from all the products

```python
df = pd.read_csv('Blinkit_Sheet2.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
monthly = df.groupby(df['order_date'].dt.to_period('M'))['order_total'].sum()
monthly.index = monthly.index.to_timestamp()
monthly.plot(title='Monthly Revenue', marker='o')
plt.ylabel('Revenue')
plt.show()
```

### 6.) Number of times each payment method done on each category analysis

```python
# correlation between payment_method and product name
data = df.pivot_table(index='payment_method', columns='category', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 10)) 
sns.heatmap(data, annot=True,fmt='d')
plt.show()
```

### 7.) Total spent analysis of each kind of user(all,inactive,newuser,premium) on each category

```python
# relation between user_status , product category and their order_total
data = df.groupby(['category', 'user_status'])['order_total'].sum().reset_index()
plt.figure(figsize=(10,10))
plt.title("Spend analysis of kinds of user on each category")
sns.lineplot(x='category', y='order_total', hue='user_status', data=data,marker='o')
plt.xticks(rotation=90)
plt.show()
```
